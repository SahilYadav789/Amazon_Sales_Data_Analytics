Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from data import clean_data

df = clean_data()
df.head(3)
```

```python
df.isnull().sum()
```

Droping Missing Values and Column

```python
# Drop critical missing values
df = df.dropna(subset=['amount'])

# Fix revenue
df['revenue'] = df['amount']

# Fill categorical missing values
df['courier_status'] = df['courier_status'].fillna('Unknown')
df['currency'] = df['currency'].fillna('INR')
df['promotion-ids'] = df['promotion-ids'].fillna('No Promotion')

# Fill location
df['ship-city'] = df['ship-city'].fillna('Unknown')
df['ship-state'] = df['ship-state'].fillna('Unknown')
df['ship-postal-code'] = df['ship-postal-code'].fillna('Unknown')
df['ship-country'] = df['ship-country'].fillna('Unknown')

# Drop weak column
df = df.drop(columns=['fulfilled-by'])

```
